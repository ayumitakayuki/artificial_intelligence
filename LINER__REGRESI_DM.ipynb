```python
import pandas as pd

# Load the Excel file to preview its structure
file_path = 'WasteManagementAndOverallRecyclingRatesAnnual.xlsx'
data = pd.ExcelFile(file_path)

# Display sheet names to understand the structure of the file
data.sheet_names
```

```python
# Load the sheet into a DataFrame to analyze its content
sheet_name = 'WasteManagementAndOverallRecycl'
df = data.parse(sheet_name)

# Display the first few rows to understand its structure
df.head()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare the data for regression
years = np.array([2018, 2019, 2020, 2021, 2022, 2023]).reshape(-1, 1)
predicted_year = np.array([2024]).reshape(-1, 1)

# Initialize dictionary to store predictions
predictions = {}

# Iterate over each data series and predict the value for 2024
for index, row in df.iterrows():
    # Extract the relevant data for regression (2018-2023)
    values = row[["2018", "2019", "2020", "2021", "2022", "2023"]].values.astype(float)

    # Fit the linear regression model
    model = LinearRegression()
    model.fit(years, values)

    # Predict for 2024
    prediction = model.predict(predicted_year)
    predictions[row["DataSeries"]] = prediction[0]

# Convert predictions to a DataFrame for better visualization
predictions_df = pd.DataFrame(list(predictions.items()), columns=["DataSeries", "2024 Prediction"])

import ace_tools as tools; tools.display_dataframe_to_user(name="Prediksi Data Tahun 2024", dataframe=predictions_df)
```
